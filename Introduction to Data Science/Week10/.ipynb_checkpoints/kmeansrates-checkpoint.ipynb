```python
import pandas as pd
```

```python
policies = pd.read_csv("Rates.csv")
```

```python
policies.head()
```

```python
import matplotlib.pyplot as plt
```

```python
pd.plotting.scatter_matrix(
    frame = policies,
    alpha = 1,
    s = 100,
    diagonal = 'none');
```

```python
X = policies.iloc[:, policies.columns != "State"]
```

```python
X.head()
```

```python
X.Gender = X.Gender.apply(lambda x: 0 if x == "Female" else 1)
```

```python
X.Gender.head()
```

```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
from sklearn.cluster import KMeans
```

```python
k_model = KMeans(
    n_clusters = 3,
    n_init = 10)
```

```python
k_model.fit(X)
```

```python
palette = {0:'#fb8072', 1:'#80b1d3', 2:'#b3de69'}
```

```python
k_colors = pd.Series(k_model.labels_) \
    .apply(lambda x:palette[x])
```

```python
pd.plotting.scatter_matrix(
    frame = policies,
    color = k_colors,
    alpha = 1,
    s = 100,
    diagonal = 'none');
```

```python
plt.scatter(
    x = policies.Age,
    y = policies.BMI,
    color = k_colors)
plt.scatter(
    x = k_model.cluster_centers_[:,5],
    y = k_model.cluster_centers_[:,4],
    marker = 'x',
    color = "black",
    s = 100)
plt.xlabel = "BMI"
plt.ylabel = "Age"
plt.show()
```
