```python
import pandas as pd
df=pd.read_csv('carsMPG_data.csv')
```

```python
df.info()
df.head()
```

# Convert columns value into one-hot-encoding

```python
dummies=pd.get_dummies(df['cylinders'])
df=pd.concat([df, dummies], axis=1)
dummies=pd.get_dummies(df['model year'])
df=pd.concat([df, dummies], axis=1)
dummies=pd.get_dummies(df['origin'])
df=pd.concat([df, dummies], axis=1)
```

```python
df.iloc[:, 9:]
```

```python
df.head(1)
```

```python
# Remove columns those we have converted into one-hot-encode
df = df.drop('cylinders', 1)
df = df.drop('model year', 1)
df = df.drop('origin', 1)
```

```python
df.head()
```

```python
X=df.iloc[:,1:] #independent feature
y=df.iloc[:,0]  #depependent feture
X=X.drop('car name',1)
```

```python
X
```

```python
y
```

```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge= Ridge()
# here we make a list of alphas from 1 to 200 and give 300 numbers in between
parameters= {'alpha': list(np.linspace(1,200, 300))}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
from sklearn.linear_model import Lasso
ridge= Lasso()
parameters= {'alpha': list(np.linspace(1,200, 300))}
lasso_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X,y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```
