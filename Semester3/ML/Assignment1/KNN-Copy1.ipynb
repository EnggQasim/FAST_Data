```python
test = pd.DataFrame({
    "Att1":[0.8,1.4,2.1],
    "Att2":[6.3,8.1,7.4],
    "k1": None,
    "k3": None,
    "k5": None,
    "k6": None
    
})
test
```

```python
def loss(actual, test):
    return (test-actual).sum()
```

# for K1

```python
k1 = pd.DataFrame(columns=["Att1",'Att2','Class','loss'])
for i in range(len(test)):
    a = test.iloc[i,0:2]
    k = train
    k['loss'] = train[["Att1","Att2"]].apply(lambda x:loss(x,a), axis=1)
    k.sort_values(by=['loss'], inplace=True)
    b = pd.DataFrame(k.iloc[0,:].copy()).T
    k1 = pd.concat([k1,b],axis=0)
    
k1    
```
