# Age Vs Different Columns Analysis

```python
cols = ['Type of Donor', 'Repeat Donor', 'HBSAg','Anti HCV', 'HIV Combo', 'RPR', 'MP']
cols
```

```python
for c in cols:
    print("Start Feature",c,"=========")
    for v in df[c].unique():
        print("Total Members of",v,"group are: ",\
             len(df[df[c]==v]),
             )
        fig, ax = plt.subplots(figsize=(15,15))
       
        df[df[c]==v].Age.value_counts().plot(kind='bar')
        for p in ax.patches:
            ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005),rotation=90)
        plt.title("Age Vs " + v)
        plt.legend()
        plt.xlabel("Ages")
        plt.ylabel("No Patients")
        plt.show()
```
