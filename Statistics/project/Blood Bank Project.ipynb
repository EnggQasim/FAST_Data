```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Blood Data for MS Course Project-2.csv")
df.describe()
```

```python
df.info()
```

# Count All Columns Unique values

```python
for c in df.columns[2::]:
    print("============Start",c,"==========")
    print(df[c].value_counts(dropna=False))
```

# Visualize Base on All Columns Unique values

```python
for c in df.columns[2::]:
    print("============Start",c,"==========")
    df[c].value_counts(dropna=False).plot('bar')
    plt.title("Population Distribution Base on "+str(c))
    plt.show()
```

# Age Vs Different Columns Analysis

```python
cols = ['Type of Donor', 'Repeat Donor', 'HBSAg','Anti HCV', 'HIV Combo', 'RPR', 'MP']
cols
```

```python
for c in cols:
    print("Start Feature",c,"=========")
    for v in df[c].unique():
        print("Total Members of",v,"group are: ",\
             len(df[df[c]==v]))
        df[df[c]==v].Age.hist()
        plt.title("Age Vs " + v)
        plt.show()
```
