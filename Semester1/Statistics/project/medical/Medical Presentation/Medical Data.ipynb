# Gender vs Disease

```python
fig, axis = plt.subplots(1,2, figsize=(10,4))
for i,g in enumerate(df.Gender.unique()):
    print(df[df.Gender==g].DIAGNOSIS.value_counts().sum())    
    df[df.Gender==g].DIAGNOSIS.value_counts().plot(kind='bar', ax=axis[i])
    axis[i].set_title(g)
    plt.suptitle("Male and Female VS Deasis", color='b', fontsize=18)
plt.show()
```

# Most common Disease in female

```python
df.Gender.value_counts()
```

```python
df[df.Gender=='Female'].DIAGNOSIS.value_counts()[:3]
```

```python
d1=df[df.Gender=='Female']
d2 = d1.DIAGNOSIS.value_counts()/d1.DIAGNOSIS.value_counts().sum()*100
d3 = d1.DIAGNOSIS.value_counts()

d2 = pd.DataFrame(d2)
d3 = pd.DataFrame(d3)

print(len(d2))
print(len(d3))

data1 = d3.merge(d2, left_index=True, right_index=True)
data1.columns = ["FemaleDiseases","F_Per"]
pd.options.display.float_format = lambda x : '{:.0f}'.format(x) if int(x) == x else '{:,.2f}'.format(x)
data1
```

```python
d1=df[df.Gender=='Male']
d2 = d1.DIAGNOSIS.value_counts()/d1.DIAGNOSIS.value_counts().sum()*100
d3 = d1.DIAGNOSIS.value_counts()

d2 = pd.DataFrame(d2)
d3 = pd.DataFrame(d3)

print(len(d2))
print(len(d3))

data2 = d3.merge(d2, left_index=True, right_index=True)
data2.columns = ["MaleDiseases","M_Per"]
pd.options.display.float_format = lambda x : '{:.0f}'.format(x) if int(x) == x else '{:,.2f}'.format(x)
data2
```

```python
data = data1.merge(data2, left_index=True, right_index=True, how='right')
data["T.Diseases"]=data.FemaleDiseases + data.MaleDiseases
data["T.Diseases_Per"]=data["T.Diseases"] / data["T.Diseases"].sum() * 100
data
```

```python

data[["FemaleDiseases","MaleDiseases"]].plot(kind='bar',title="Male Vs Females")


plt.xlabel("Diseases")
plt.yticks([0,400,1500])
plt.ylabel("Volume")
```

# 5. In which year female having most common Disease 

```python
years = df[(df.Gender=='Female')].YEAR.value_counts().index

for y in years:
    print("Year: ", y)
    print("Total",df[(df.YEAR==y) & (df.Gender=='Female')]['DIAGNOSIS'].value_counts().sum())
    print(df[(df.YEAR==y) & (df.Gender=='Female')]['DIAGNOSIS'].value_counts()[:3])
    plt.show()
```

```python
fig, axis = plt.subplots(1,5, figsize=(20,4))
for i,y in enumerate(years):
    d1 =df[(df.Gender=="Female") & (df.YEAR==y)]
    print(d1.DIAGNOSIS.value_counts().sum())    
    d1.DIAGNOSIS.value_counts().plot(kind='bar', ax=axis[i])
    axis[i].set_xlabel(y)
    
plt.suptitle("Year and Female VS Deasis\n\n\n", color='b', fontsize=28)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter

# Fixing random state for reproducibility
np.random.seed(19680801)
```

```python
fig, axs = plt.subplots(1, 2, tight_layout=True)

# N is the count in each bin, bins is the lower-limit of the bin
N, bins, patches = axs[0].hist(df.Age.values, bins=10)

# We'll color code by height, but you could use any scalar
fracs = N / N.max()

# we need to normalize the data to 0..1 for the full range of the colormap
norm = colors.Normalize(fracs.min(), fracs.max())

# Now, we'll loop through our objects and set the color of each accordingly
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)

# We can also normalize our inputs by the total number of counts
axs[1].hist(df.Age.values, bins=10, density=True)

# Now we format the y-axis to display percentage
axs[1].yaxis.set_major_formatter(PercentFormatter(xmax=1))
```
