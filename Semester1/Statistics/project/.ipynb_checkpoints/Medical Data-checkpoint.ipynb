```python
fig, axs = plt.subplots(1, 2, tight_layout=True)

# N is the count in each bin, bins is the lower-limit of the bin
N, bins, patches = axs[0].hist(df.Age.values, bins=10)

# We'll color code by height, but you could use any scalar
fracs = N / N.max()

# we need to normalize the data to 0..1 for the full range of the colormap
norm = colors.Normalize(fracs.min(), fracs.max())

# Now, we'll loop through our objects and set the color of each accordingly
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)

# We can also normalize our inputs by the total number of counts
axs[1].hist(df.Age.values, bins=10, density=True)

# Now we format the y-axis to display percentage
axs[1].yaxis.set_major_formatter(PercentFormatter(xmax=1))
```

# Solve Classification problem

```python
import pandas as pd
df = pd.read_excel("./medical/Medical Data for MS Course Project (Research).xlsx")
df.columns = ['sno','Age','Gender','DIAGNOSIS','YEAR']
df.dtypes
```

```python
df.head()
```

```python
gender_map={v:k for k, v in list(enumerate(df.Gender.unique()))}
diagnosis_map={v:k for k, v in list(enumerate(df.DIAGNOSIS.unique()))}
print(gender_map)
print(diagnosis_map)
print(len(diagnosis_map))
```

```python
df1 = df.iloc[:]

df1['Gender'] = df1['Gender'].map(gender_map)

df1['DIAGNOSIS'] = df1['DIAGNOSIS'].map(diagnosis_map)
df1.head()
```

```python
df1.columns
```

```python
import numpy as np
for c in ['sno', 'Age', 'Gender', 'YEAR']:
    df1[c] = df1[c].astype('float32')
    
df1.dtypes    
```

```python
d=int(len(df1)*70/100)
print(d)
```

```python
x_train = df1.iloc[:d,[0,1,2,4]]
x_test = df1.iloc[d:,[0,1,2,4]]

y_train = df1.iloc[:d, 3]
y_test = df1.iloc[d:,3 ]

y_train
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
network = models.Sequential()
network.add(layers.Dense(120, activation='relu', input_shape=(4,)))
network.add(layers.Dense(20, activation='softmax'))
```

```python
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

```python
network.fit(x_train, y_train, epochs=20, batch_size=128)
```

```python
network.evaluate(x_test, y_test)
```

# Random Forest

```python
#Importing Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.externals import joblib
print('Libraries Imported')
```

```python
import pandas as pd
df = pd.read_excel("./medical/Medical Data for MS Course Project (Research).xlsx")
df.columns = ['sno','Age','Gender','DIAGNOSIS','YEAR']
df.dtypes
```

```python
df.columns
```

```python
#Creating the dependent variable class
factor = pd.factorize(df['DIAGNOSIS'])
df.DIAGNOSIS = factor[0]
definitions = factor[1]
print(df.DIAGNOSIS.head())
print(definitions)
```

```python
#Creating the dependent variable class
factor1 = pd.factorize(df['Gender'])
df.Gender = factor1[0]
df.head(2)
```

```python
x = df[['Age', 'Gender', 'YEAR']].values
y = df[['DIAGNOSIS']].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 21)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Fitting Random Forest Classification to the Training set
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
print(list(zip(['Age', 'Gender', 'YEAR'], classifier.feature_importances_)))
joblib.dump(classifier, 'randomforestmodel.pkl')
```
