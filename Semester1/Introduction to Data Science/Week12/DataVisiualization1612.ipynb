# Suplots

```python
import matplotlib.pyplot as plt
import numpy as np
time_1 = np.arange(12)
revenue = time_1 * 200
costs= time_1 *300
fig, axis = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
plt.title("Revenue VS Cost")

axis[0].plot(time_1, revenue, marker='o')
axis[0].set_title("Cost Vs Time")
axis[0].set_xlabel("Time")
axis[0].set_ylabel("Revenue")

axis[1].plot(time_1, costs, marker='X')
axis[1].set_xlabel("Time")
axis[1].set_ylabel("Cost")

plt.show()
```

```python
import matplotlib.pyplot as plt
time_1 = [0,1,2,3,4]
revenue = [200,400,650,800,850]
costs = [150,500,550,550,650]
plt.title("Revenue VS Cost")
plt.axis([0,5,10,800])
plt.plot(time_1, revenue, marker='o')
plt.plot(time_1, costs, marker='X')
plt.xlabel("Time")
plt.ylabel("Valume")
plt.legend(['Revenue','Cost'], loc=1)

plt.show()
```
